### Question 1

Calculons l'espérance:
\begin{align*}
\mathbb{E}_{\theta}[X_1] 
&= \int_{-\infty}^{+\infty} x \cdot \frac{\mathbf{1}_{[0, \theta]}(x)}{\theta} \, dx\\
&= \int_0^{\theta} x \frac{1}{\theta} \, dx \\
&= \frac{1}{\theta} \int_0^{\theta} x \, dx \\
&= \frac{1}{\theta} \cdot \frac{\theta^2}{2} \\
&= \frac{\theta}{2}
\end{align*}

Par la méthode des moments :
$$
\frac{1}{n} \sum_{i=0}^{n} X_i = \overline X = \frac{\widehat{\theta}_1}{2}
$$


### Question 2

Calculons le risque lié à l'estimateur 
$$
R(\widehat{\theta}_1,\theta)= \operatorname{Var}(\widehat{\theta}_1) + \operatorname{Biais}(\theta_1,\theta)^2
$$
Calul de la Variance de Xi 

\begin{align*}
\mathbb{E}_{\theta}[X_1^2]
&= \int_{-\infty}^{+\infty} x^2 \cdot \frac{\mathbf{1}_{[0, \theta]}(x)}{\theta} \, dx\\
&= \int_0^{\theta} x^2 \frac{1}{\theta} \, dx \\
&= \frac{1}{\theta} \int_0^{\theta} x^2 \, dx \\
&= \frac{1}{\theta} \cdot \frac{\theta^3}{3} \\
&= \frac{\theta^2}{3}
\end{align*}

Ainsi : 
\begin{align*}
\operatorname{Var}(\widehat{\theta}_1)
&= 4\operatorname{Var}(\overline X) \\
&= 4\cdot\frac{\operatorname{Var}(X_1)}{n^2} \\
&= \frac{4}{n^2}\cdot(\mathbb{E}_{\theta}[X_1^2] - \mathbb{E}_{\theta}[X_1]^2) \\
&= \frac{4}{n^2}\cdot(\frac{\theta^2}{3} - \frac{\theta^2}{4}) \\
&= 4\cdot\frac{\theta^2}{12n} \\
&=\frac{\theta^2}{3n}
\end{align*}


Calcul du Biais: 
\begin{align*}
\operatorname{Biais}(\widehat{\theta}_1,\theta) 
&= \mathbb{E}[\widehat{\theta}_1]- \theta \\
&=2\mathbb{E}[\overline X]- \theta \\
&=\frac{2}{n}\cdot \sum_{i=0}^{n} \mathbb{E}[X_i] - \theta \\
&=\frac{2}{n}\cdot\frac{n \cdot \theta}{2} - \theta \\
&=0
\end{align*}

Ainsi
$$ R(\widehat{\theta}_1,\theta)=\frac{\theta^2}{3n}. $$

### Question 3

On a 
$$ L(\theta)=\prod_{i=1}^n p_\theta(x) = \prod_{i=1}^n \frac{1}{\theta}\mathbf{1}_{[0,\theta]}(X_i)=\theta^{-n}\mathbf{1}_{\{\max X_{i}\le\theta\}}. $$

On voit que L est décroissante sur $\mathopen{[}\max_{1\le i\le n}X_i\,;+\infty\mathclose{[}$ et vaut $0$ partout ailleurs, donc L est maximisée lorsque $\theta=\max_{1\le i\le n}X_i$.
$$ \widehat{\theta}_2= \widehat{\theta}_{MLE} = \operatorname*{argmax}_{\theta \in \Theta}(L(\theta)) = \max_{1\le i\le n}X_i. $$

### Question 4

Pour $x\in[0,\theta]$,

$$
F_{\widehat{\theta}_2}(x)=\mathbb{P}(\widehat{\theta}_2\le x)=\mathbb{P}(X_1\le x,\dots,X_n\le x)
=\prod_{i=1}^n\mathbb{P}(X_i\le x)=\bigl(F(x)\bigr)^n=\Bigl(\frac{x}{\theta}\Bigr)^n.
$$

$$F_{\widehat{\theta}_2}(x) = \begin{cases}
0, & \text{if } x < 0, \\
(\frac{x}{\theta})^n, & \text{if } 0 \le x \le \theta, \\
1, & \text{if } x > \theta
\end{cases}$$

En dérivant sur $(0,\theta)$ on obtient la densité :
$$
f_{\widehat{\theta}_2}(x)=\frac{d}{dx}F_{\widehat{\theta}_2}(x)=n\frac{x^{n-1}}{\theta^n},\qquad 0<x<\theta.
$$
Ainsi la dénsité de probabilté associé à $\widehat{\theta}_2$
$$
p_{\widehat{\theta}_2}(x)=n\frac{x^{n-1}}{\theta^n} \mathbf{1}_{[0,\theta]}
$$
On a donc :
\begin{align*}
\mathbb{E}[\widehat{\theta}_2]
&= \int_{-\infty}^{+\infty} x \cdot n\frac{x^{n-1}}{\theta^n} \mathbf{1}_{[0,\theta]} \, dx\\
&= \frac{n}{\theta^n}\int_0^{\theta} x^n\,dx \\
&= \frac{n}{n+1}\,\theta, 
\end{align*}
et
$$
\mathbb{E}[\widehat{\theta}_2^2]
=\frac{n}{\theta^n}\int_0^{\theta} x^{n+1}\,dx
=\frac{n}{n+2}\,\theta^2.
$$
Calculons la Variance de $\widehat{\theta}_2$
$$
\operatorname{Var}(\widehat{\theta}_2)=\left(\frac{n}{n+2}-\frac{n^2}{(n+1)^2}\right)\theta^2
$$
Calculons le Biais entre de $\widehat{\theta}_2$ et $\theta$
$$
\operatorname{Bias}(\widehat{\theta}_2,\theta)=\frac{n}{n+1}\,\theta-\theta=-\frac{\theta}{n+1},
$$
Nous pouvons maintenant caluculer le risque:
\begin{align*}
R(\widehat{\theta}_2,\theta)
&=\left(\frac{n}{n+2}-\frac{n^2}{(n+1)^2}\right)\theta^2+\frac{\theta^2}{(n+1)^2} \\
&= \left(\frac{n}{n+2}-\frac{n^2-1}{(n+1)^2}\right)\theta^2 \\
&= \left(\frac{n}{n+2}-\frac{(n+1)(n-1)}{(n+1)^2}\right)\theta^2 \\
&= \left(\frac{n}{n+2}-\frac{n-1}{n+1}\right)\theta^2 \\
&= \left(\frac{n(n+1) - (n-1)(n+2)}{(n+2)(n+1)}\right)\theta^2 \\
&=\frac{2\theta^2}{(n+1)(n+2)}.
\end{align*}


### Question 5

Nous avons par la question précéndente et par la question 2:
$$ R(\widehat{\theta}_1,\theta)=\frac{\theta^2}{3n},\qquad R(\widehat{\theta}_2,\theta)=\frac{2\,\theta^2}{(n+1)(n+2)}. $$
$R(\widehat{\theta}_2)$ décroît en $O(\frac{1}{n^2})$ tandis que $R(\widehat{\theta}_1)=O(\frac{1}{n})$.

Donc pour $n$ assez grand (nombres échantillons assez grandes), le risque de $\widehat{\theta}_2$ devient negligable face au risque de $\widehat{\theta}_1$.


Plus le risque est faible meilleur seras l'estimateur au sens du MSE. De plus, on remarque même que $R(\widehat{\theta}_2)\le R(\widehat{\theta}_1)$ pour $n \ge 2$, donc clairement $\widehat{\theta}_2$ est un meilleur estimateur pour tout n.

## Exercice 3: Learning a linear classifier

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
```

### Question 1

Generate ample set in R^2 and choosing an arbitrary normal vector w∗ to define the separating hyperplane of the two classes. The sign of ⟨w∗, xi⟩ gives the label yi ∈ {−1, +1}. (2pt)

```python
def h(x, w):
    return x.dot(w)
```

```python
np.random.seed(0)

# Number of data points
N = 500

# Generate Random Data in R2
X = np.random.randn(N, 2)
X_bar = np.hstack((np.ones((N, 1)),X))

#Chose arbitarty line for classification
w_star = np.random.randn(3)
w_star = w_star / np.linalg.norm(w_star)

slope = -w_star[1] / w_star[2]
intercept = -w_star[0] / w_star[2]

# Label data according to which side of the line they are on
y = np.sign(h(X_bar, w_star)).astype(int)
y[y == 0] = 1  # Handle the rare case where a point lies exactly on the line
assert np.all(np.unique(y) == np.array([-1, 1]))

# Plot the data
plt.figure(figsize=(6, 6))
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class +1')
plt.scatter(X[y == -1, 0], X[y == -1, 1], color='red', label='Class -1')
plt.title("Randomly Generated Data with Labels")
plt.legend()
plt.xlabel("X1")
plt.ylabel("X2")
plt.axline((0, intercept), slope=slope, color='red', linestyle='--')
plt.show()
```

### Question 2

$$
\begin{array}{l}
\textbf{Algorithm: Stochastic Gradient Descent (SGD) for Adaline model} \\[6pt]

\textbf{Given: } \text{training data } \{(x_i, y_i)\}_{i=1}^n,\ \text{learning rate } \eta > 0, \text{ and epochs } T \\[6pt]

\textbf{Model: } h(\bar{x}; \bar{w}) = \langle \bar{w}, \bar{x} \rangle,\ \text{where } \bar{x} = (x, 1),\ \bar{w} = (w, \tau) \\[6pt]

\textbf{Empirical Loss: } 
L_n(\bar{w}) = \frac{1}{n} \sum_{i=1}^{n} (y_i - h(\bar{x}_i; \bar{w}))^2 \\[8pt]

\textbf{for } t = 1 \text{ to } T \textbf{ do} \\[3pt]
\quad \text{Randomly sample all training examples } \{(x_i, y_i)\}_{i=1}^n \text{ without replacement} \\[3pt]
\quad \text{for each sampled example } (x_i, y_i) \text{ do} \\[3pt]
\qquad \text{Compute prediction: } \hat{y}_i = h(\bar{x}_i; \bar{w}_t) = \langle \bar{w}_t, \bar{x}_i \rangle \\[3pt]
\qquad \text{Compute gradient: } g_i = -2 (y_i - \hat{y}_i) \bar{x}_i \\[3pt]
\qquad \text{Update parameters: } \bar{w}_{t+1} = \bar{w}_t - \eta \, g_i \\[3pt]
\quad \text{end for} \\[3pt]
\textbf{end for} \\[6pt]

\textbf{Output: } \bar{w}_T \text{ (final parameter vector)} 
\end{array}
$$


### Question 3

```python
# Creation of X_bar with bias term
X_bar = np.hstack((np.ones((N, 1)), X))  # Add a column of ones for the bias term
# X_bar shape is (N, 3)
```

```python
# Adaline MSE loss 
def adaline_loss(w, x, y):
    err = y - h(x, w)
    return np.mean(err ** 2)

# Adaline Gradient computation 
def adaline_gradient(w, x, y):
    grad = -2 * (y - h(x, w)) * x
    return grad
    
```

```python
# 1. Define model
def adaline_sgd(Xb, y, lr=0.01, epochs=20):
    n, d = Xb.shape
    w = np.zeros(d)
    losses = []

    for epoch in range(epochs):
        # Shuffle indices each epoch
        indices = np.random.permutation(n)
        for i in indices:
            xi = Xb[i]
            yi = y[i]
            w -= lr * adaline_gradient(w, xi, yi)
            # Compute loss after each update
            losses.append(adaline_loss(w, Xb, y))

    return w, losses
```

```python
# 2. Train model
w_sgd, losses = adaline_sgd(X_bar, y, lr=0.02, epochs=1)

print("Learned weights (including bias):", w_sgd)
print("Final MSE:", losses[-1])
```

```python
# 3. Evaluate classification accuracy
y_pred = np.sign(X_bar @ w_sgd)
acc = np.mean(y_pred == y)
print(f"Classification accuracy: {acc*100:.2f}%")
```

```python
# 4. Plot results
tau = w_sgd[0]
w1 = w_sgd[1]
w2 = w_sgd[2]

slope_sgd = -w1 / w2
intercept_sgd = -tau / w2

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(losses)
plt.xlabel("Step")
plt.ylabel("MSE")
plt.title("MSE ")

plt.subplot(1,2,2)

slope_sdg = -w_sgd[1] / w_sgd[2]
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class +1')
plt.scatter(X[y == -1, 0], X[y == -1, 1], color='red', label='Class -1')
plt.title("Randomly Generated Data with Labels")
plt.xlabel("X1")
plt.ylabel("X2")
plt.axline((0, intercept), slope=slope, color='red', linestyle='--',label='True Decision Boundary')
plt.axline((0, intercept_sgd), slope=slope_sgd, color='green', linestyle='--', label='Adaline Decision Boundary')
plt.legend()
plt.show()

```

```python
# Compare with original line 
print("Norm of the error between w_star and w_est: ", np.linalg.norm(w_sgd - w_star))
```

L'estimation $\hat{w}$ est assez proche de $w^\star$ (erreur quadratique plus petite que 1).

```python
# Learnt bias and weights
bias = w_sgd[0]
weights = w_sgd[1:]
print("Learned bias:", bias)
print("Learned weights:", weights)
# True weights were w_star with no bias term
print("True weights:", w_star)
```
